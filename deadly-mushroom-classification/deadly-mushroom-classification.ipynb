## Deadly Mushroom Classification

Data source: https://www.kaggle.com/datasets/uciml/mushroom-classification

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv("mushrooms.csv")
data
```

### Preprocessing

```python
data.columns
```

```python
mappings = list()

encoder = LabelEncoder()

for i in range(len(data.columns)):
    data[data.columns[i]] = encoder.fit_transform(data[data.columns[i]])
    mappings_dict = {index: label for index, label in enumerate(encoder.classes_)}
    mappings.append(mappings_dict)
```

```python
mappings
```

```python
y = data['class']
X = data.drop('class', axis=1)
```

```python
y
```

```python
X
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=123)
```

### Model Selection

```python
log_model = LogisticRegression()
svm_model = SVC(C=1.0, kernel='rbf')
nn_model = MLPClassifier(hidden_layer_sizes=(128, 128))
```

### Training

```python
np.sum(y) / len(y)
```

```python
log_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
nn_model.fit(X_train, y_train)
```

```python
print(f"   Logistic Regression: {log_model.score(X_test, y_test)}")
print(f"Support Vector Machine: {svm_model.score(X_test, y_test)}")
print(f"       Neural Networks: {nn_model.score(X_test, y_test)}")
```
